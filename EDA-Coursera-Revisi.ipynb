# Pre-processing and Cleaning Data 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"E:\Data Analyst\Datasets\Coursera Courses\coursera_courses.csv")
df.head()
```

```python
df = df.drop('course_url', axis=1)
df.head()
```

## Dealing with Null

```python
df.isna().sum()
```

```python
df = df.dropna()
df
```

```python
df.isna().sum()
```

## Data type

### REMOVING K FROM COURSE REVIEWS NUM AND CONVERT THEM TO 1000( 3.6K - 3600)

```python
df["course_reviews_num"]
```

```python
df['course_reviews_num'] = df['course_reviews_num'].replace(r'[k]', '', regex=True).astype(float) * 1000
```

```python
df['course_reviews_num'] = df['course_reviews_num'].astype(int)
```

```python
df["course_reviews_num"]
```

```python
df.info()
```

### course_students_enrolled and course_reviews_num

```python
df['course_students_enrolled'] = df['course_students_enrolled'].str.replace(',','')
```

```python
df['course_students_enrolled'] = df['course_students_enrolled'].astype('int64')
df['course_reviews_num'] = df['course_reviews_num'].astype('int64')
```

```python
df.info()
```

### convert course_time into int

```python
df['course_time'].unique()
```

```python
def convert_to_hours(category):
    if 'Less Than 2 Hours' in category:
        return 2  # Misalkan kita anggap sebagai 2 jam
    elif '1 - 4 Weeks' in category:
        return 4 * 7 * 24  # Misalkan kita anggap sebagai 4 minggu
    elif '1 - 3 Months' in category:
        return 3 * 4 * 7 * 24  # Misalkan kita anggap sebagai 3 bulan
    elif '3 - 6 Months' in category:
        return 6 * 4 * 7 * 24  # Misalkan kita anggap sebagai 6 bulan
    else:
        return 0

df['course_time_hours'] = df['course_time'].apply(convert_to_hours)
df
```

### convert course difficulty into int

```python
df['course_difficulty'].unique()
```

```python
difficulty_mapping = {'Beginner': 1, 'Intermediate': 2, 'Mixed': 3, 'Advanced': 4}
df['course_difficulty_numeric'] = df['course_difficulty'].map(difficulty_mapping)
df
```

### convert course certificate type into int

```python
df['course_certificate_type'].unique()
```

```python
#Di urutkan berdasarkan tingkat kesulitannya
certificate_mapping = {
    'Course': 1,
    'Guided Project': 2,
    'Professional Certificate': 3,
    'Specialization': 4
}

df['course_certificate_numeric'] = df['course_certificate_type'].map(certificate_mapping)
df
```

### The most common skills on coursera

```python
from collections import Counter
import ast
```

```python
course_skills = df['course_skills']
all_skills = ', '.join(course_skills)
all_skills
```

```python
cleaned_data = ast.literal_eval(all_skills)
cleaned_data
```

```python
individual_skills = [skill for sublist in cleaned_data for skill in sublist]
individual_skills
```

```python
skill_counts = Counter(individual_skills)
```

```python
skill_counts
```

```python
skill_counts = pd.DataFrame(list(skill_counts.items()), columns=['Skill', 'Count'])
skill_counts
```

```python
skill_counts = skill_counts.sort_values(by='Count', ascending=False)
skill_counts
```

```python
most_common_skills = skill_counts.head(5)
most_common_skills
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Skill', data=most_common_skills, palette='viridis')
plt.title('Top 5 Most Common Skills')
plt.xlabel('Count')
plt.ylabel('Skill')
plt.show()
```

```python
skill_counts.to_excel('E:\Data Analyst\Datasets\Coursera Courses\skill_counts.xlsx', index=False)
```
